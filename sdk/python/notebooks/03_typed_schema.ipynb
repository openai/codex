# Typed + Schema wrappers
Use typed wrappers to reduce key-lookup mistakes.

```python
from codex_app_server import AppServerClient

with AppServerClient() as client:
    client.initialize()
    started = client.thread_start_typed(model="gpt-5")
    print(started.thread.id)

    schema_started = client.thread_start_schema(model="gpt-5")
    print(schema_started.thread.id)
```
